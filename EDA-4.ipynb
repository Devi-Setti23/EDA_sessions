# pie chart
- pie chart is a 360 dgrees view
- it is under matplotlib package
- it will take keys and values
- that we will get for value counts

```python
# step-1:get keys from value counts
# step-2:get values from value counts
#plt.pie verify the arguments
# which arguments takes values and keys
```

```python
cnf=visa_df['continent'].value_counts()
cnf
```

```python
x=cnf.values
x
```

```python
y=cnf.keys()
y
```

```python
plt.pie(x,labels=y,autopct='%0.2f%%')
plt.show()
```
