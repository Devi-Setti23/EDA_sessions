# fit and transform

- fit means develop the logic
- transform means process the logic
- if we observe in map first we develop dictionary logic
- then we apply the dictionary to column
- developing the dictionary is kind of fit
- apply the dictionary to column means we are transforming the data cat to num cols

**np.where**
- using np.where also we can change the categorical data to numerical 
- but np.where is only for binary condition
- np.where is same as if and else
- if condition is true then we apply true value
- if the condition is false we apply false value
- so np.where is use for only binary labels
- a column should have only 2 unique labels

```python
# value=='certified'=====  replace with 0===otherwise 1
visa_df=pd.read_csv(r"C:\Users\devis\OneDrive\Documents\data science\data files\Visadataset.csv")
```

**one hot encoder**

- one hot means one will be ON, anther will be OFF
- ON reprsents with 1
- OFF represents with 0
- for example case status has 2 unique labels 
  - certified
  - denied
- one hot encoder creates new column which is equal to no of unique labels
- for example case_status will create two new extra columns
    - case_status_certified
    - case_status_denied

- case_status     case_status_certified      case_status_denied

  denied                0                           1
  certified             1                           0
  denied                0                           1

**advantages**
- using one hot encoder we will ge new columns
- the new column values at a time only one will be **True** another will be **False**
- which means new columns are **independent each other**
- independent means **no relation**
- means no **multi collineraity**
    -multi collineraity is an imp property in ML
- **machine learning models expect the columns should be independent each other**
- so one hot encoder gives the no multicollineraity
    


**disadvantages**
  
- the disadv is it will increase many columns
- in the above example the total cols becomes 25510
- this is happen because of case id
- in real time we will drop the id columns
- eventhough if we drop still we are getting 30 cols
- more columns means more time to process
- we required more infrastructure(CPU-GPU-TPU)
- this is called as **curse od dimensionality**
