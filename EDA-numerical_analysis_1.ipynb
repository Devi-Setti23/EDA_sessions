# finding the outliers


- we already know tat outliers available less than Q1-1.5*IQR and Q3+1.5*IQR
- step-1:
     - calculate Q1=25p
    - calculate Q2=50p
    - calculate Q3=75p
- step-2:calculate IQR=Q3-Q1
- step-3: 
       - calculate lb=Q1-1.5*IQR
      - CAalculate ub=Q3+1.5*IQR
- step-4:
   - con1=wage data<lb
   - con2=wate data>ub
   - con3=con1|con2
- step-5:data[con3]

# how to deal outliers
- method-1:drop the outliers
   - droping outliers is easy task but not recommended
   - if you drop the outliers means we are droping the rows
   - which means we are droping the other column values also
   - may be some important we might loose from other column values 
   - but in industry if outlier percentage <2% of the total data we can drop
   - the number of outliers in pwage is 427
   - the number of total observations are 25480
   - the percentage is 1.67 which is less than 2%

- method-2: fill with median value

   - WKT outlier doesnt effect by median
   - so we fill the outliers with median


- method-3: capping or florring

  - instead of fill with median we can fill with certain values
  - the outliers which are less than lb fill with lb
  - the outliers which are greater than ub fill with ub value only

- isolation forest

  - ML model is used to find the outliers and anamolies
